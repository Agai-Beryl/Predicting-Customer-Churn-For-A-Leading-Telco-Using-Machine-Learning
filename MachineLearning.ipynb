### One Hot Encoding Categorical Variables

To use in classification models, categorical features are transformed into dummy variables with values of 0 and 1.


We will create a class instance `OneHotLabelEncoder` that encodes our  categorical variables.

We have excluded `phone number` because we already dropped it and also `Churn` because it will be our target variable.

We are left with:

- International plan

- Voicemail plan

`State`  and `Area code` features will be  encoded separately as written below.


### Label encoding the State and Area Code feature

The `state` and `area code` column is transformed using the LabelEncoder technique, which assigns a unique integer to each distinct label in the column. This encoding method is chosen over creating dummy variables due to the large number of distinct values in the "state" column.

Using dummy variables in this case could potentially introduce issues with the interpretation of feature importance, especially in tree-based models. By applying LabelEncoder, each state label is represented by a corresponding integer, preserving the integrity of the column while avoiding the complications that may arise from using dummy variables.

```python
#Instantiating the class and initializing it 

class DataEncoder:
    def __init__(self, dataset):
        self.dataset = dataset
    
    def one_hot_encode(self):
        # Select categorical columns for one-hot encoding
        categorical_columns = ['international plan', 'voice mail plan']
        
        # Perform one-hot encoding for categorical columns
        one_hot_encoded_features = pd.get_dummies(self.dataset[categorical_columns])
        return one_hot_encoded_features
    
    def label_encode(self, column):
        # Initialize LabelEncoder for a specific column
        label_encoder = LabelEncoder()
        
        # Fit label encoder and transform the specified column
        self.dataset[column] = label_encoder.fit_transform(self.dataset[column])
        return self.dataset
    
    def encode_dataset(self):
        # Perform one-hot encoding
        one_hot_encoded_features = self.one_hot_encode()
        
        # Perform label encoding for 'state' and 'area code'
        self.label_encode('state')
        self.label_encode('area code')
        
        # Concatenate one-hot encoded features with the original dataset
        # Drop the original categorical columns that were one-hot encoded and label encoded to avoid redundancy
        final_encoded_dataset = pd.concat([self.dataset.drop(columns=['international plan', 'voice mail plan']), one_hot_encoded_features], axis=1)
        
        return final_encoded_dataset

#Calling it to our dataset
encoder = DataEncoder(dataset)
final_encoded_dataset = encoder.encode_dataset()

# Display final encoded dataset
final_encoded_dataset.head()
```

```python
#Instantiate class

class CorrelationHeatmap:
    def __init__(self, data):
        self.data = data
    
    def plot_heatmap(self):
        # Compute the correlation matrix
        corr_matrix = self.data.corr()
        
        # Plot the heatmap
        plt.figure(figsize=(15, 12))
        sns.heatmap(corr_matrix, annot=True, cmap='YlOrRd', fmt='.2f', linewidths=0.5)
        plt.title('Correlation Heatmap', fontsize=16)
        plt.xlabel('Features', fontsize=12)
        plt.ylabel('Features', fontsize=12)
        plt.xticks(rotation=90, ha='right')
        plt.yticks(rotation=0)
        plt.tight_layout()
        plt.show()

#Calling it to our dataset
heatmap_generator = CorrelationHeatmap(final_encoded_dataset)
heatmap_generator.plot_heatmap()

```

However, some features in the dataset exhibit perfect positive correlation, such as "Total day charge" and "Total day minutes", "Total eve charge" and "Total eve minutes", "Total night charge" and "Total night minutes", and "Total int charge" and "Total int minutes".

Highly correlated variables should be considered for dropping due to redundancy, which can lead to overfitting and reduced interpretability in predictive models. Excluding them improves computational efficiency and facilitates clearer interpretation of variable effects on the target outcome. However, caution is advised to ensure the retention of essential information and context-specific considerations.

#### Dropping Highly-Correlated Features

Dropping features that have a correlation of 0.9 or above.


```python
def drop_highly_correlated_features(data, threshold=0.9):
    # Compute the correlation matrix
    corr_matrix = data.corr().abs()
    
    # Select upper triangle of correlation matrix
    upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))
    
    # Find features with correlation greater than or equal to the threshold
    to_drop = [column for column in upper.columns if any(upper[column] >= threshold)]
    
    # Drop highly correlated features
    data_dropped = data.drop(columns=to_drop)
    
    return data_dropped

#Calling it to our dataset
final_dataset_no_high_correlation = drop_highly_correlated_features(final_encoded_dataset, threshold=0.9)

final_dataset_no_high_correlation.shape
```
